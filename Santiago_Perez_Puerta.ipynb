```python
# %% Anatomia de una grafica, tomado de http://matplotlib.org
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import AutoMinorLocator, MultipleLocator, FuncFormatter

np.random.seed(19680801)

X = np.linspace(0.5, 3.5, 100)
Y1 = 3+np.cos(X)
Y2 = 1+np.cos(1+X/0.75)/2
Y3 = np.random.uniform(Y1, Y2, len(X))

fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(1, 1, 1, aspect=1)


def minor_tick(x, pos):
    if not x % 1.0:
        return ""
    return "%.2f" % x

ax.xaxis.set_major_locator(MultipleLocator(1.000))
ax.xaxis.set_minor_locator(AutoMinorLocator(4))
ax.yaxis.set_major_locator(MultipleLocator(1.000))
ax.yaxis.set_minor_locator(AutoMinorLocator(4))
ax.xaxis.set_minor_formatter(FuncFormatter(minor_tick))

ax.set_xlim(0, 4)
ax.set_ylim(0, 4)

ax.tick_params(which='major', width=1.0)
ax.tick_params(which='major', length=10)
ax.tick_params(which='minor', width=1.0, labelsize=10)
ax.tick_params(which='minor', length=5, labelsize=10, labelcolor='0.25')

#------------------------------------------------------------------------------------
ax.grid(linestyle="-", linewidth=1, color='.25', zorder=-10,which="major")
ax.grid(linestyle="--", linewidth=0.5, color='.25', zorder=-10,which="minor")

ax.plot(X, Y1, c=(0.25, 0.25, 1.00), lw=2, label="Blue signal", zorder=10)
ax.plot(X, Y2, c=(1.00, 0.25, 0.25), lw=2, label="Red signal")
ax.plot(X, Y3, linewidth=0,
        marker='o', markerfacecolor='w', markeredgecolor='k')

ax.set_title("Anatomy of a figure", fontsize=20, verticalalignment='bottom')
ax.set_xlabel("X axis label")
ax.set_ylabel("Y axis label")

ax.legend()


def circle(x, y, radius=0.15):
    from matplotlib.patches import Circle
    from matplotlib.patheffects import withStroke
    circle = Circle((x, y), radius, clip_on=False, zorder=10, linewidth=1,
                    edgecolor='black', facecolor=(0, 0, 0, .0125),
                    path_effects=[withStroke(linewidth=5, foreground='w')])
    ax.add_artist(circle)


def text(x, y, text):
    ax.text(x, y, text, backgroundcolor="white",
            ha='center', va='top', weight='bold', color='blue')


# Minor tick
circle(0.50, -0.10)
text(0.50, -0.32, "Minor tick label")

# Major tick
circle(-0.03, 4.00)
text(0.03, 3.80, "Major tick")

# Minor tick
circle(0.00, 3.50)
text(0.00, 3.30, "Minor tick")

# Major tick label
circle(-0.15, 3.00)
text(-0.15, 2.80, "Major tick label")

# X Label
circle(1.80, -0.27)
text(1.80, -0.45, "X axis label")

# Y Label
circle(-0.27, 1.80)
text(-0.27, 1.6, "Y axis label")

# Title
circle(1.60, 4.13)
text(1.60, 3.93, "Title")

# Blue plot
circle(1.75, 2.80)
text(1.75, 2.60, "Line\n(line plot)")

# Red plot
circle(1.20, 0.60)
text(1.20, 0.40, "Line\n(line plot)")

# Scatter plot
circle(3.20, 1.75)
text(3.20, 1.55, "Markers\n(scatter plot)")

# Grid
circle(3.00, 3.00)
text(3.00, 2.80, "Grid")

# Legend
circle(3.70, 3.80)
text(3.70, 3.60, "Legend")

# Axes
circle(0.5, 0.5)
text(0.5, 0.3, "Axes")

# Figure
circle(-0.3, 0.65)
text(-0.3, 0.45, "Figure")

color = 'blue'
ax.annotate('Spines', xy=(4.0, 0.35), xytext=(3.3, 0.5),
            weight='bold', color=color,
            arrowprops=dict(arrowstyle='->',
                            connectionstyle="arc3",
                            color=color))

ax.annotate('', xy=(3.15, 0.0), xytext=(3.45, 0.45),
            weight='bold', color=color,
            arrowprops=dict(arrowstyle='->',
                            connectionstyle="arc3",
                            color=color))

ax.text(4.0, -0.4, "Made with http://matplotlib.org",
        fontsize=10, ha="right", color='.5')

plt.show()
```

```python
# %% Programa Prueva tecnica
import numpy as np

n = int(input("Ingrese el valor de n:\n"))
a = list(map(int,input("\nIngrese el valor de los dos pares de coordenadas (x,y), separadas por espacio:\n").strip().split()))

PuntosX = []
PuntosY = []
aux = 1
aux1 = 1

if (a[2]- a[0])<0 or (a[3] - a[1]) < 0:
  aux = a[0]
  a[0]= a[2]
  a[2]=aux

  aux1=a[1]
  a[1]=a[3]
  a[3]=aux1

  aux = -1
  aux1 = 1

mat = np.ones((n,n))
deltaX = abs(a[2] - a[0])
deltaY = abs(a[3] - a[1])
costo = deltaX + deltaY -1

for i in range(1,deltaX+1):
  PuntosX.append((a[0]+i,a[1]))

if deltaX != 0:
  for i in range(1,deltaY):
    PuntosY.append((PuntosX[-1][0],a[1]+i))
else:
  for i in range(1,deltaY+1):
    PuntosY.append((a[0],a[1]+i))

if deltaX == 0:
  del(PuntosY[-1])
elif deltaY == 0:
  del(PuntosX[-1])

Recorrido = PuntosX+PuntosY

for i in Recorrido:
    mat[i]= 0

print(mat)
print("Puntos ingresados: {0}".format(a))
print("DeltaX = {0}".format(deltaX))
print("DeltaY = {0}".format(deltaY))
print("Costo = {0}".format(costo))
print("PuntosX: {0}".format(PuntosX))
print("PuntosY: {0}".format(PuntosY))
print("Recorrido: {0}".format(Recorrido[::aux][::aux1]))
```

```python
def f(h):

  #Librearías necesaria
  from numpy import pi #Importa el vlaor de pi de la libraría numpy

  #Declaración de variables internas
  R= 3.3 #Establece R como 3.3
  V=33 #Establece R como 33

  return pi*h**3-3*pi*R*h**2+3*V #Retorna el valor calculado de f(h)

h = float(input('Ingrese el valor de h que deea evaluar:\n'))
print(f'f(h)={f(h):0.3f}')
```

```python
# %% Definir los precios de los artículos en un diccionario
precios = {'A': 120, 'B': 250, 'C': 360}

# Inicializar el total de la compra
total_compra = 0

# Bucle para solicitar claves de artículos y calcular el total
while True:
    clave = input("Ingrese la clave del artículo (A, B, C) o 'X' para finalizar:\n")# Lee el dato de entrada

    clave = clave.upper()  # Convertir a mayúsculas para ser case-insensitive

    if clave == 'X':#Si el dato ingresado es iguala X

        break  # Salir del bucle si se ingresa 'X'

    if clave in precios: # Verificar si el dato ingresado es valido para el problema

        precio_articulo = precios[clave]# Asignar el precio del articulo pedido a la variable precio_articulo

        total_compra += precio_articulo# Calcular el total de la compra (esto se hace en cada ciclo)

        print(f"Precio del artículo {clave}: {precio_articulo}\n")# Mostrar en pantalla el precio del artículo

    else: #Si el valor ingresado no es valido...

        print("Clave inválida. Ingrese una clave válida (A, B, C) o 'X' para finalizar.")#Imprimir mensaje de error

# Mostrar el total de la compra
print(f"Total de la compra: {total_compra}")
```

```python
import random  # Importar la biblioteca random para generar números aleatorios

# Definir la función para generar y mostrar 50 números aleatorios entre 0 y 70
def aleatorios1():
    print("Números aleatorios entre 0 y 70:")
    for _ in range(50):
        num = random.randint(0, 70)  # Generar número aleatorio entre 0 y 70
        print(f'{num}', end=' ')  # Mostrar número en la misma línea
    print('\n')  # Imprimir línea en blanco para separar

# Definir la función para generar y mostrar 30 números aleatorios entre 30 y 80
def aleatorios2():
    print("Números aleatorios entre 30 y 80:")
    for _ in range(30):
        num = random.randint(30, 80)  # Generar número aleatorio entre 30 y 80
        print(f'{num}', end=' ')  # Mostrar número en la misma línea
    print('\n')  # Imprimir línea en blanco para separar

# Definir la función del menú
def menu():
    print("Menú:")
    print("1. Generar números aleatorios entre 0 y 70")
    print("2. Generar números aleatorios entre 30 y 80")
    print("3. Salir")
    opcion = int(input("Selecciona una opción: "))
    return opcion

# Definir la función main
def main():
    opcion = menu()
    while opcion != 3:  # Mientras la opción no sea salir
        if opcion == 1:
            aleatorios1()  # Llamar a la función aleatorios1
        elif opcion == 2:
            aleatorios2()  # Llamar a la función aleatorios2
        else:
            print("Opción inválida. Por favor, selecciona una opción válida.")
        opcion = menu()  # Volver a mostrar el menú y obtener nueva opción

# Llamar a la función main para iniciar el programa
main()
```

```python
# %% El encriptador de David 🦒
def encriptador_David(palabra):
  alfabeto_latino = 'abcdefghijklmnñopqrstuvwxyz'

  palabra_encriptada = palabra[0]

  for letra in palabra[1:]:
    posicion_letra_alfabeto = alfabeto_latino.index(letra)
    posicion_encriptada = posicion_letra_alfabeto + palabra.index(letra)
    largo_alfabeto = len(alfabeto_latino)

    if (posicion_letra_alfabeto != largo_alfabeto -1) and (posicion_encriptada <= largo_alfabeto -1):
      palabra_encriptada += alfabeto_latino[posicion_encriptada]

  return palabra_encriptada
#-------------------------------------------------------------------------------

# %% main
frase = input('Escriba la frase a encriptar:\n').lower().replace('.','').replace(',','').split(' ')
frase_encriptada = ''

for palabra in frase:
  if len(palabra) > 1:
    frase_encriptada += ' '+ encriptador_David(palabra)
  else:
    frase_encriptada += ' '+palabra

frase_encriptada = frase_encriptada[1:]

print(f'\n{frase_encriptada}')
```

```python
# %% Calculador de factores primos
def primo(numero:int)-> bool:
  '''
  Esta función verifica si un numero es primo
  '''

  if numero < 2: return False

  raiz_de_numero = int((numero)**0.5)+1

  for n in range(2,raiz_de_numero):
      if numero % n == 0:
        return False
  return True

def generador_de_primos(numero:int) -> list:
  '''
  Devuelve una lista con todos los numeros primos menores
  o iguales que el parametro Ingresado.
  '''

  numeros_primos = []

  for i in range(numero+1):
    if primo(i):
      numeros_primos.append(i)

  return numeros_primos

def calcular_factores_primos(numero:int) -> list:
  '''
  Calcula los factores primos del numero ingresado como parametro.
  '''
  factores_primos = []

  if numero == 0: return 'El cero no tiene factores primos.'

  numeros_primos = generador_de_primos(numero)

  if numero == 1:
    return numero

  else:
    factor = 2

    while numero != 1:
      if (factor in numeros_primos) and (numero % factor == 0):
        factores_primos.append(factor)
        numero = numero/factor
      else:
        factor += 1

  return factores_primos

from numpy import prod
from random import randint

numero = randint(0,100)
a = calcular_factores_primos(numero)

print(f'Estos son los factores primos de {numero}:\n{a}\n\nComrpocavión:{prod(a)}')
```
