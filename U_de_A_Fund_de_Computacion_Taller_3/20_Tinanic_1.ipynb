```python
# Libreías
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme()
```

```python
# Cantidad de variables y registros
titanic_df = pd.read_csv('titanic.csv', index_col= 0)
```

```python
filas,columnas = titanic_df.shape
print(f'Registros: {filas}\nVariables: {columnas}')
```

```python
# MODIFICACIÓN DEL FORMATO DE LOS LOS TÍTULOS DE LAS COLUMNAS
titanic_df.columns = titanic_df.columns.str.strip().str.lower().str.replace(' ','_')
titanic_df.columns
```

```python
# Info del Df
titanic_df.info()
```

```python
# Primeras 10 filas
titanic_df.head(10)
```

```python
# Ultimas 10 filas 
titanic_df.tail(10)
```

```python
# Pasajero con el ID 148
titanic_df.loc[148]
```

```python
# Filas pares
titanic_df.loc[titanic_df.index % 2 == 0]
```

```python
# Primera clase ordenada alfabéticamente según su nombre
titanic_df.name[titanic_df.pclass == 1].sort_values()
```

```python
# Porcentaje de supervivencia y mortalidad
titanic_df.survived.value_counts(normalize=True)
```

```python
# Hay edades nulas ? 
titanic_df.age.isnull().any()
```

```python
# Confirma que ya no hay edades nulas
titanic_df.dropna(subset='age', inplace=True)
titanic_df.age.isnull().any()
```

```python
# Mayor de edad? 
titanic_df['mayor_de_edad'] = titanic_df.age >= 18
titanic_df.sample(10)
```

```python
titanic_df[['mayor_de_edad','pclass','survived']].groupby(['pclass','mayor_de_edad',]).value_counts(normalize=True)
```

```python
titanic_df['survived'].value_counts(normalize=True).plot(kind='pie', autopct='%1.1f%%', labels= ['Fallecidos', 'Sobrevivientes'])
plt.title('Tasa de mortalidad y supervivencia')
plt.ylabel('')
plt.show()
```

```python
titanic_df['age'].plot(kind='hist')
plt.title('Histograma de Edades')
plt.xlabel('Edad')
plt.grid()
plt.show()
```

```python
titanic_df[['pclass', 'survived']].groupby(['pclass', 'survived']).size().unstack().plot(kind='bar')
plt.xticks(rotation = 0)
plt.xlabel('Clases')
plt.ylabel('Numero de sobrevivientes')
plt.show()
```

```python
titanic_df[['pclass', 'survived']].groupby(['pclass', 'survived']).size().unstack().plot(kind='bar', stacked= True)
plt.xticks(rotation = 0)
plt.xlabel('Clases')
plt.ylabel('Numero de sobrevivientes')
plt.show()
```
